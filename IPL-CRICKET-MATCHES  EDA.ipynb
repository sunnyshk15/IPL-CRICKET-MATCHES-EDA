Number of matches played in each IPL season

```python
plt.figure(figsize=(10, 6))
season_counts = df['Season'].value_counts().sort_index()
season_counts.plot(kind='bar', color='green')

plt.title('No. of Matches Played in each IPL season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=80)
plt.show()
```

Matches won by the Teams

```python
plt.figure(figsize=(10,6))
winner_counts=df['winner'].value_counts().sort_index()
winner_counts.plot(kind='bar',color='blue')
plt.title('Number of Wins by Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins()')
plt.xticks(rotation=80)
plt.show()
```

No of matches played in stadium

```python
plt.figure(figsize=(15,8))
match_count=df['venue'].value_counts().sort_index()
match_count.plot(kind='bar',color='purple')
plt.title('No. of Matches Played in stadium')
plt.xlabel('stadium')
plt.ylabel('No.of Matches')
plt.show()
```

Toss Decision to field or bat 

```python
sns.countplot(data=df,x='toss_decision')
plt.title('Total Toss Decision to field or bat')
plt.xlabel('toss_decision')
plt.ylabel('count')
plt.xticks(rotation=75)
plt.show()
```

Toss Decision to Field or Bat in each ipl season 

```python
plt.figure(figsize=(15,6))
sns.countplot(data=df,x='Season',order=df['Season'].value_counts().index.sort_values(),hue='toss_decision')
plt.title('Toss Decision to Field or Bat in each ipl season')
plt.xlabel('season')
plt.ylabel('count')
plt.xticks(rotation=75)
plt.show()
```

Player of the Match

```python
plt.figure(figsize=(10,6))
player_count=df['player_of_match'].value_counts().iloc[:10]
player_count.plot(kind='bar',color='blue')
plt.title('player of the Match')
plt.xlabel('player')
plt.ylabel('count')
plt.xticks(rotation=75)
plt.show()
```
